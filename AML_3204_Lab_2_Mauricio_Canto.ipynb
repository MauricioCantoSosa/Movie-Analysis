```python
import pandas as pd
import re
```

```python
f = open('/content/YouTubeTrailerComments.txt', 'rb')
```

```python
Movieids = []
descriptors = []
df = pd.DataFrame(columns = ['MovieID', 'View Count', 'Likes Count', 'Dislikes Count', 'Forward/Share Count', 'Commnets Count'])
```

```python
dfrow = 0
info_row = []
while True:
        line = f.readline().decode('ISO-8859-1')
        if not line:
            break
        if 'MovieID::::' in line:
          if len(info_row) == 1 and len(df) > 2:
            falses = [False, False, False, False, False]
            info_row.extend(falses)
            df.loc[len(df)] = info_row
            info_row = []
          id = re.search(r'\d+', line)
          info_row.append(id.group())  
        if re.match(r'\d+\s\d+\s\d+\s\d+\s\d+', line):
          row = line.split()
          info_row.extend(row)
          df.loc[len(df)] = info_row
          info_row = []
```
